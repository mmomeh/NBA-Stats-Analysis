```python
#loading data
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
pd.set_option('display.max_columns', None)

data = pd.read_excel('nba_player_data.xlsx')
```

```python
#sample of data
data.sample(10)
```

```python
#dimensions of the data
data.shape
```

  Data  cleaning and analysis preparation

```python
#checking for missing values
data.isna().sum()
```

```python
#dropping unnecessary columns
data.drop(columns = ['RANK', 'EFF'], inplace=True)
```

```python
#extracting the season start year
data['season_start_year'] = data['Year'].str[:4].astype(int)
```

```python
#replacing team name abbreviations
data['TEAM'].replace(to_replace=['NOP','NOH'], value=['NO','NOK'], inplace=True)
```

```python
#replacing season type abbreviation
data['Season_type'].replace('Regular%20Season', 'RS', inplace=True)
```

```python
# splitting data into Regular Season and Playoffs
rs_df = data[data['Season_type']=='Regular Season']
playoffs_df = data[data['Season_type']=='Playoffs']
```

```python
#column names
data.columns
```

```python
#defining list of total columns
total_cols = ['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA',
       'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']
```

Which player stats are correlated with each other?

```python
#calculating correlation matrix
correlation_matrix = data[total_cols].corr()
display(correlation_matrix)
```

```python
# Analyzing per-minute player stats and their correlations
data_per_min = data.groupby(['PLAYER', 'PLAYER_ID', 'Year'])[total_cols].sum().reset_index()
for col in total_cols: # iterate through total_cols instead of data_per_min.columns[4:]
    data_per_min[col] = data_per_min[col] / data_per_min['MIN']

#shooting %
data_per_min['FG%'] = data_per_min['FGM']/data_per_min['FGA']
#3pt %
data_per_min['3PT%'] = data_per_min['FG3M']/data_per_min['FG3A']
#freethrow %
data_per_min['FT%'] = data_per_min['FTM']/data_per_min['FTA']
#% of 3pt attempts
data_per_min['FG3A%'] = data_per_min['FG3A']/data_per_min['FGA']
#average pts per attempt
data_per_min['PTS/FGA'] = data_per_min['PTS']/data_per_min['FGA']
#% of 3pt makes
data_per_min['FG3M/FGM'] = data_per_min['FG3M']/data_per_min['FGM']
#freethrow attempt per field goal attempt
data_per_min['FTA/FGA'] = data_per_min['FTA']/data_per_min['FGA']
#true shooting %
data_per_min['TRU%'] = 0.5*data_per_min['PTS']/(data_per_min['FGA']+0.475*data_per_min['FTA'])
#turnover ratio
data_per_min['AST/TOV'] = data_per_min['AST']/data_per_min['TOV']

#filter out players who do not play as much
data_per_min = data_per_min[data_per_min['MIN']>=50]

# Drop non-numeric columns before calculating correlation
data_per_min.drop(columns=['PLAYER', 'PLAYER_ID', 'Year'], inplace=True)

fig = px.imshow(data_per_min.corr())
fig.show()
```

How are minutes played distributed?

```python
# Plotting dist of minutes played
fig = px.histogram(x=playoffs_df['MIN'], histnorm = 'percent')
fig.show()
```

```python
#Helper function to calculate ppg
def hist_data(df=rs_df, min_MIN= 0, min_GP= 0) :
  return df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'PTS']/\
  df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'GP']
```

```python
#Comparing ppg dist for Regular Season and Playoffs
fig = go.Figure()
fig.add_trace(go.Histogram(x=hist_data(rs_df, 50, 5), histnorm= 'percent', name= 'Regular Season',
                           xbins={'start': 0, 'end': 38, 'size' : 1}))
fig.add_trace(go.Histogram(x=hist_data(playoffs_df, 5, 1), histnorm= 'percent', name= 'Playoffs',
                           xbins={'start': 0, 'end': 38, 'size' : 1}))
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.50)
fig.show()
```

```python
#calculating percentage of playoff players in a scoring range
((hist_data(playoffs_df,5,1)>=12) & (hist_data(playoffs_df,5,1)<=34)).mean()
```

How has the game changed over the past 10 yers?

```python
# Calculating season stats and percentages
change_df = data.groupby('season_start_year')[total_cols].sum().reset_index()
change_df['POSS_est'] = change_df['FGA'] - change_df['OREB'] + change_df['TOV'] + 0.44*change_df['FTA']
change_df = change_df[list(change_df.columns[0:2])+ ['POSS_est']+ list(change_df.columns[2:-1])]

#shooting %
change_df['FG%'] = change_df['FGM']/change_df['FGA']
#3pt %
change_df['3PT%'] = change_df['FG3M']/change_df['FG3A']
#freethrow %
change_df['FT%'] = change_df['FTM']/change_df['FTA']
#% of 3pt attempts
change_df['FG3A%'] = change_df['FG3A']/change_df['FGA']
#average pts per attempt
change_df['PTS/FGA'] = change_df['PTS']/change_df['FGA']
#% of 3pt makes
change_df['FG3M/FGM'] = change_df['FG3M']/change_df['FGM']
#freethrow attempt per field goal attempt
change_df['FTA/FGA'] = change_df['FTA']/change_df['FGA']
#true shooting %
change_df['TRU%'] = 0.5*change_df['PTS']/(change_df['FGA']+0.475*change_df['FTA'])
#turnover ratio
change_df['AST/TOV'] = change_df['AST']/change_df['TOV']
```

```python
# Plotting yearly average per-team stats per 48 minutes
# Calculate Per-Team, Per-48 Minute Stats
team_stats = data.groupby(['Year', 'TEAM'])[total_cols].sum().reset_index()
team_minutes = data.groupby(['Year', 'TEAM'])['MIN'].sum().reset_index(name='TEAM_MIN')
team_stats = team_stats.merge(team_minutes, on=['Year', 'TEAM'])

for col in total_cols:
    team_stats[f'{col}_per48_team'] = (team_stats[col] / team_stats['TEAM_MIN']) * 48
    team_stats[f'{col}_per48_team'] = team_stats[f'{col}_per48_team'].replace([float('inf'), float('-inf')], 0).fillna(0)

# Calculate Yearly Average Per-Team Stats
yearly_avg_per48_team = team_stats.groupby('Year').mean(numeric_only=True).reset_index()
cols_to_drop = total_cols + ['TEAM_MIN', 'MIN']
cols_to_drop_present = [col for col in cols_to_drop if col in yearly_avg_per48_team.columns and not col.endswith('_per48_team')]
yearly_avg_per48_team.drop(columns=cols_to_drop_present, inplace=True)

# Plot Yearly Average Per-Team Stats
fig = go.Figure()
for col in yearly_avg_per48_team.columns[1:]:
    # Remove '_per48_team' suffix from the legend label
    legend_name = col.replace('_per48_team', '')
    fig.add_trace(go.Scatter(x=yearly_avg_per48_team['Year'], y=yearly_avg_per48_team[col], mode='lines+markers', name=legend_name))

fig.update_layout(
    title='Yearly Average Per-Team Statistics Per 48 Minutes', # Add title
    xaxis_title='Year',
    yaxis_title='Average Statistic Per 48 Minutes Per Team',
    hovermode='x unified'
)

fig.show()
```

```python
# plotting yearly average per-team stats per 100 possessions
change_per100_df = change_df.copy()
#calculate per 100 possessions for relevant columns
for col in change_per100_df.columns[2:18] : # Assuming columns 2 to 17 are the relevant stats
  change_per100_df[col] = (change_per100_df[col]/change_per100_df['POSS_est'])*100

#drop 'MIN' and 'POSS_est' columns
change_per100_df.drop(columns=['MIN', 'POSS_est'], inplace=True)

fig = go.Figure()
#iterate through columns of change_per100_df for plotting
for col in change_per100_df.columns[1:]:
  fig.add_trace(go.Scatter(x=change_per100_df['season_start_year'], y=change_per100_df[col], name=col))

fig.show()
```

Compare Regular Season to Playoffs

```python
rs_change_df = rs_df.groupby('season_start_year')[total_cols].sum().reset_index()
playoffs_change_df = playoffs_df.groupby('season_start_year')[total_cols].sum().reset_index()

for i in [rs_change_df, playoffs_change_df] :
  i['POSS_est'] = i['FGA'] - i['OREB'] + i['TOV'] + 0.44*i['FTA']
  i['POSS_per_48'] = (i['POSS_est']/i['MIN'])*48

  #shooting %
  i['FG%'] = i['FGM']/i['FGA']
  #3pt %
  i['3PT%'] = i['FG3M']/i['FG3A']
  #freethrow %
  i['FT%'] = i['FTM']/i['FTA']
  #% of 3pt attempts
  i['FG3A%'] = i['FG3A']/i['FGA']
  #average pts per attempt
  i['PTS/FGA'] = i['PTS']/i['FGA']
  #% of 3pt makes
  i['FG3M/FGM'] = i['FG3M']/i['FGM']
  #freethrow attempt per field goal attempt
  i['FTA/FGA'] = i['FTA']/i['FGA']
  #true shooting %
  i['TRU%'] = 0.5*i['PTS']/(i['FGA']+0.475*i['FTA'])
  #turnover ratio
  i['AST/TOV'] = i['AST']/i['TOV']

  for col in total_cols :
    i[col] = 100 * i[col]/i['POSS_est']
  i.drop(columns=['MIN','POSS_est'], inplace=True)

rs_change_df
```

```python
#calculating percentage change between Playoffs and Regular Season
comp_change_df = round(100*(playoffs_change_df-rs_change_df)/rs_change_df,3)
comp_change_df['season_start_year'] = list(range(2015,2024))
comp_change_df
```

```python
#plotting percentage change between Playoffs and Regular Season
fig = go.Figure()

for col in comp_change_df.columns[1:] :
  fig.add_trace(go.Scatter(x=comp_change_df['season_start_year'], y=comp_change_df[col], name=col))

fig.show()
```

```python
#calculate Per-Player, Per-48 Minute Stats
player_stats = data.groupby(['PLAYER', 'PLAYER_ID', 'Year', 'Season_type'])[total_cols].sum().reset_index()

#calculate per-48 minute stats
# Handle division by zero for players with 0 minutes played
for col in total_cols:
    player_stats[f'{col}_per48'] = (player_stats[col] / player_stats['MIN']) * 48
    # Replace infinite values (resulting from division by zero) with 0
    player_stats[f'{col}_per48'] = player_stats[f'{col}_per48'].replace([float('inf'), float('-inf')], 0)
    # Replace NaN values (resulting from division by zero) with 0
    player_stats[f'{col}_per48'] = player_stats[f'{col}_per48'].fillna(0)


#display the first few rows of the resulting DataFrame
display(player_stats.head())
```

```python
#group by year,calculate the mean of the per-48 stats, excluding non-numeric columns
yearly_avg_per48 = player_stats.groupby('Year').mean(numeric_only=True).reset_index()

#drop unnecessary columns
cols_to_drop = total_cols + ['PLAYER_ID']
#drop columns that are part of the original total_cols or PLAYER_ID
cols_to_drop_present = [col for col in cols_to_drop if col in yearly_avg_per48.columns and not col.endswith('_per48')]
yearly_avg_per48.drop(columns=cols_to_drop_present, inplace=True)

#display the first few rows of the yearly averages
display(yearly_avg_per48.head())
```

```python
#plot Yearly Averages
fig = go.Figure()

#iterate through the columns (excluding Year) to add to the plot
for col in yearly_avg_per48.columns[1:]:
    fig.add_trace(go.Scatter(x=yearly_avg_per48['Year'], y=yearly_avg_per48[col], mode='lines+markers', name=col))
#update layout
fig.update_layout(
    title='Yearly Average Per-Player Statistics Per 48 Minutes',
    xaxis_title='Year',
    yaxis_title='Average Statistic Per 48 Minutes',
    hovermode='x unified'
)

fig.show()
```
